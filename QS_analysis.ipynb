```python
#I'm putting the date object back in. 
def fixYear(x):
    try:
        y = datetime.strptime(x, '%m/%d/%Y')
    except:
        y = datetime.strptime(x, '%m/%d/%y')
    return y

all_data['dataObje'] = all_data.index
all_data['dataObje'] = all_data['dataObje'].apply(lambda X: fixYear(X))
#replacing the index
all_data.index = all_data['dataObje']
```

```python
mysubset = ['Going from work to home','Out for drinks']

#couple of functions to convert Series objects to arrasy that work better with matplotlib
def getX_Y(df,mysubset):
    '''x,y = getX_Y(mergedDF,mysubset)'''
    subset = df.dropna(0,subset=mysubset)[mysubset]
    x = subset[mysubset].loc[:,mysubset[0]].tolist()
    y = subset[mysubset].loc[:,mysubset[1]].tolist()
    return np.asarray(x),np.asarray(y)

def getX(df,col):
    '''x,y = getX_Y(df,col)'''
    subset = df.dropna(0,subset=[col])
    x = subset[col].tolist()
    return np.asarray(x)

#and a couple of solid matplotlib graphing functions:
def compareTwoArrays(x,y,color='#253B68',kind='hex'):
    '''kind : { scatter | reg | resid | kde | hex }
     stat_func : {spearmanr | pearsonr }
     hue="species"
     '''
    sns.set_palette([color])
    sns.set_context("notebook",font_scale=2.0)
    with sns.axes_style("white"):
        sns.jointplot(x,y,kind=kind,size=16,stat_func=None);
        
def singleVarHistogram(arr,bins=10,title="histogram",xaxis="value"):
    plt.hist(arr,bins)
    plt.title(title)
    plt.xlabel(xaxis)
    plt.ylabel("Frequency")
    
def bunchOfVarLine(arrays,labels,label="chart"):
    '''This is for when you just want to bang out a quick graph of several variables'''
    ind = range(len(arrays))
    [plt.plot(arrays[x],label=labels[x]) for x in ind]
    plt.ylabel(label)
    #plt.xticks(range(len(arrays[0])+1))
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```

## Benchmarking and looking for 'excessive drinking'

I'm using the <a href="http://www.cdc.gov/alcohol/fact-sheets/alcohol-use.htm">CDC guidlines for excessive drinking</a>

```python
#counting the number of drinks per day
all_data['hadDrinks'] = all_data['Out for drinks']>0
drinksdf = all_data[all_data['hadDrinks']]
#i'm using the assumption that a drink costs 10 dollars rounded up. 
#This means that a $20 tab is equal to two drinks.
#As these reciepts probaby include some food, this is a conservative estimate of actual alchohol consumption
days_drinks = all_data['numDrinks'] = all_data['Out for drinks'].apply(lambda x: math.ceil(x/10)).dropna()
singleVarHistogram(np.array(days_drinks.tolist()),bins=30,title="Distribution of drinks",xaxis="#")
```

```python
days_drinks.describe()
```

Among nights where I drink any alcohol, the most common is one to three drinks. Most of the 'Out for drinks' nights involve 1 to 10 drinks, with a few really crazy evenings.

When I go out, I have an average of three and a half drinks. The max of 28 is unlikely and probably includes food and a show.

```python
days_drinks.quantile(.68)
```

```python
days_drinks.quantile(.95)
```

But let's look at the total days, not just days that involve 'going out'.

```python
#same as before but .fillna allows me to include days where there are no 'out for drinks' items.
days_drink2 = all_data['numDrinks'] = all_data['Out for drinks'].apply(lambda x: math.ceil(x/10)).fillna(0)
```

Up to the first standard deviation, I'm likely to drink no more than two drinks. That's actually within CDC guidelines. I can live with that.

Looking at the other metric: no more than 15 drinks per week. This is more of a problem. With a limit of only two drinks per day it's easy to go over. Some weeks I go out a lot and then again on the weekend. 

```python
drinksByWeek = pd.DataFrame(days_drinks.resample('W',how='sum'))
drinksByWeek['limit'] = 15
drinksByWeek['control'] = drinksByWeek['Out for drinks'] - drinksByWeek['limit']
drinksByWeek['over'] = drinksByWeek['control'] > 0

bunchOfVarLine([np.array(drinksByWeek['Out for drinks']),
                np.array(drinksByWeek['limit'])],
                ['drinks','limit'],
                label="Total drinks")
```

```python
len(drinksByWeek[drinksByWeek['over']])/float((len(drinksByWeek)))
```

According to my metrics. I go over this limit almost a quarter of the time. This is clearly the metric to monitor in the future.
